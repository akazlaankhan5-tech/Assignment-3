```python
# Arithmetic Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)

# Comparison Operators
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)

# Assignment Operators
c = 5
c += 2
print("+= :", c)
c -= 1
print("-= :", c)
c *= 2
print("*= :", c)
c /= 2
print("/= :", c)

# Logical Operators
x = True
y = False

print("AND:", x and y)
print("OR:", x or y)
print("NOT:", not x)

# Bitwise Operators
print("Bitwise AND:", a & b)
print("Bitwise OR:", a | b)
print("Bitwise XOR:", a ^ b)
print("Bitwise NOT:", ~a)
print("Left Shift:", a << 1)
print("Right Shift:", a >> 1)

# Membership Operators
list1 = [1, 2, 3, 4]

print("2 in list:", 2 in list1)
print("5 not in list:", 5 not in list1)

# Identity Operators
m = [1, 2]
n = m
p = [1, 2]

print("m is n:", m is n)
print("m is p:", m is p)
print("m is not p:", m is not p)
```

```python
# Marks obtained
english = 85
islamiat = 90
maths = 95

# Total marks
total_marks = 300

# Obtained marks
obtained_marks = english + islamiat + maths

# Calculate percentage
percentage = (obtained_marks / total_marks) * 100

# Display results
print("English Marks:", english)
print("Islamiat Marks:", islamiat)
print("Maths Marks:", maths)
print("Total Marks:", total_marks)
print("Obtained Marks:", obtained_marks)
print("Percentage:", percentage, "%")
```
